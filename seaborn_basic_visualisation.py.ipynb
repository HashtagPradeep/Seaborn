```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

```

```python
rawdata=pd.read_csv('D:\git_code\Marketing_Raw_Data.csv')

print(rawdata.shape)

rawdata.head()
```

```python
# 1. LINE GRAPH
```

```python
axis= sns.lineplot(x='Week_ID', y= 'Revenue', data=rawdata,
                   hue= 'Promo',style = 'Promo', ci = 68, markers= True)

# Visualising the revenue over weeks with hue color of RGB and style to the 
# lines 
```

```python
# Increase the size
axis= sns.lineplot(x='Week_ID', y= 'Revenue', data=rawdata,
                   hue= 'Promo',style = 'Promo', ci = None)
sns.set(rc={'figure.figsize':(12,10)})
```

```python
# add markers to the line plot
axis= sns.lineplot(x='Month_ID', y= 'Revenue', data=rawdata,
                   hue= 'Promo',style = 'Promo', ci = None, markers=True)
```

```python
# BY day names
axis= sns.lineplot(x='Month_ID', y= 'Revenue', data=rawdata,
                   hue= 'Day_Name',style = 'Promo', ci = None, markers=True)
sns.set(rc={'figure.figsize':(14,12)})
```

```python
# By year
axis= sns.lineplot(x='Year', y= 'Revenue', data=rawdata,
                   hue= 'Day_Name',style = 'Promo', ci = None, markers=True)
```

```python
#(i) Total Revenue by Month

axis = sns.barplot(x= 'Month_ID',y='Revenue', data = rawdata)

rawdata[['Month_ID','Revenue']].groupby('Month_ID', as_index = False).agg({'Revenue':'mean'}).head()
```
